```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers=pd.read_csv('Customers.csv')
```

```python
customers.head()
```

```python
customers.isnull().sum()
```

```python
products=pd.read_csv("Products.csv")
```

```python
transactions=pd.read_csv("Transactions.csv")
```

```python
products.isnull().sum()
```

```python
products.isnull().sum()
```

```python
transactions.head()
```

```python
df=transactions.merge(customers,on='CustomerID',how = 'left')
df=df.merge(products,on='ProductID',how='left')
```

```python
df.head()
```

```python
df
```

```python
df.drop('Price_x',axis=1)
```

```python
df.rename(columns={'Price_y':'Price'})
```

```python
df.head()
```

```python
df.drop('Price_y',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'Price_x':'Price'},inplace=True)
```

```python
df.head()
```

```python
total_revenue=df['Price'].sum()
```

```python
total_revenue
```

```python
df['TransactionDate']=pd.to_datetime(df['TransactionDate'])
```

```python
df.head()
```

```python
transaction_over_time=df.groupby(df['TransactionDate'].dt.to_period('M')).size()
```

```python
transaction_over_time
```

```python
transaction_over_time.plot(kind='line',title='Transaction Over Time',figsize=(10,5))
plt.xlabel("Month")
plt.ylabel("Number of Transaction")
plt.show()
```

```python
region_analysis=df.groupby('Region')['Price'].agg(['sum','count']).reset_index()
```

```python
sns.barplot(data=region_analysis,x='Region',y='sum')
plt.title('Revenue by region')
plt.xlabel('Region')
plt.ylabel("Total Revenue")
plt.show()
```

# Product Performance


```python
top_products=df.groupby("ProductName")['Price'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar',title='Top selling Products',figsize=(10,5))

plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
